```python
#Plotting a Simple Graph

import numpy as np

import matplotlib.pyplot as plt

xvalues = np.array([0, 1, 2, 3, 4, 5])

yvalues = np.array([0, 2, 1, 4, 3, 6])

other_yvalues = np.array([1, 3, 1, 3, 5, 5])

xerrors = np.array([0.5, 0.3, 0.2, 0.1, 0.4, 0.2])

yerrors = np.array([0.35, 0.4, 0.25, 0.4, 0.3, 0.45])

other_yerrors = np.array([0.45, 0.3, 0.4, 0.25, 0.4, 0.35])

import scipy.optimize as scpo 

def Line1(x, gradient1, intercept1): 
    
    return gradient1*x + intercept1 

actual_fit_parameters, covariance_matrix = scpo.curve_fit(Line1, xvalues,yvalues) 

fit_gradient1 = actual_fit_parameters[0] 

fit_intercept1 = actual_fit_parameters[1] 

ybestfit1 = Line1(xvalues, fit_gradient1, fit_intercept1) 

def Line2(x, gradient2, intercept2): 
    
    return gradient2*x + intercept2 

actual_fit_parameters, covariance_matrix = scpo.curve_fit(Line2, xvalues,other_yvalues) 

fit_gradient2 = actual_fit_parameters[0] 

fit_intercept2 = actual_fit_parameters[1] 

ybestfit2 = Line2(xvalues, fit_gradient2, fit_intercept2)

plt.scatter(xvalues, yvalues, color="blue", label="First Data Points")
plt.plot(xvalues, ybestfit1, color="blue")

plt.scatter(xvalues, other_yvalues, color="red", label="Second Data Points")
plt.plot(xvalues, ybestfit2, color="red")

plt.errorbar(xvalues, yvalues, xerr=xerrors, yerr=yerrors, fmt='o', color="blue", elinewidth=1, capsize=3)

plt.errorbar(xvalues, other_yvalues, xerr = xerrors , yerr = other_yerrors , fmt = 'o', color='red', elinewidth=1, capsize=3)

plt.gca().set_xlim([-0.7,5.7])

plt.gca().set_ylim([-0.7,6.7])

plt.xlabel("x values")

plt.ylabel("y values")

plt.legend()

plt.show()
```

```python
#Plotting a Simple Graph with Residuals

import numpy as np

import matplotlib.pyplot as plt

xvalues = np.array([0, 1, 2, 3, 4, 5])

yvalues = np.array([0, 2, 1, 4, 3, 6])

other_yvalues = np.array([1, 3, 1, 3, 5, 5])

xerrors = np.array([0.5, 0.3, 0.2, 0.1, 0.4, 0.2])

yerrors = np.array([0.35, 0.4, 0.25, 0.4, 0.3, 0.45])

other_yerrors = np.array([0.45, 0.3, 0.4, 0.25, 0.4, 0.35])


import scipy.optimize as scpo 

def Line1(x, gradient1, intercept1): 
    
    return gradient1*x + intercept1 

actual_fit_parameters, covariance_matrix = scpo.curve_fit(Line1, xvalues,yvalues) 

fit_gradient1 = actual_fit_parameters[0] 

fit_intercept1 = actual_fit_parameters[1] 

ybestfit1 = Line1(xvalues, fit_gradient1, fit_intercept1) 

def Line2(x, gradient2, intercept2): 
    
    return gradient2*x + intercept2 

actual_fit_parameters, covariance_matrix = scpo.curve_fit(Line2, xvalues,other_yvalues) 

fit_gradient2 = actual_fit_parameters[0] 

fit_intercept2 = actual_fit_parameters[1] 

ybestfit2 = Line2(xvalues, fit_gradient2, fit_intercept2)


residuals1 = yvalues - ybestfit1
residuals2 = other_yvalues - ybestfit2


fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True, gridspec_kw={'height_ratios':[3,1]},figsize=(7,6))


ax1.scatter(xvalues, yvalues, color="blue", label="First Data Points")
ax1.plot(xvalues, ybestfit1, color="blue")

ax1.scatter(xvalues, other_yvalues, color="red", label="Second Data Points")
ax1.plot(xvalues, ybestfit2, color="red")

ax1.errorbar(xvalues, yvalues, xerr=xerrors, yerr=yerrors,fmt='o', color="blue", elinewidth=1, capsize=3)
ax1.errorbar(xvalues, other_yvalues, xerr=xerrors, yerr=other_yerrors,fmt='o', color="red", elinewidth=1, capsize=3)

ax1.set_xlim([-0.7,5.7])
ax1.set_ylim([-0.7,6.7])
ax1.set_ylabel("y values")
ax1.legend()


ax2.errorbar(xvalues, residuals1, xerr=xerrors, yerr=yerrors, fmt='o', color="blue", capsize=3)
ax2.errorbar(xvalues, residuals2, xerr=xerrors, yerr=other_yerrors, fmt='o', color="red", capsize=3)

ax2.axhline(0, color="black", linestyle="--")
ax2.set_xlabel("x values")
ax2.set_ylabel("Residuals")

plt.tight_layout()
plt.show()
```
